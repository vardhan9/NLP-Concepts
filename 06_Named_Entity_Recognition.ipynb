```python
sentence = "India is incredible country, one of the oldest civilization and got independence on 15th august 1947"
```

```python
import nltk
words = nltk.word_tokenize(sentence)
words
```

```python
# pos tag
tag_elements = nltk.pos_tag(words)
tag_elements
```

```python
nltk.download('all') # used all because unable to find maxent_ne_chunker
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```
